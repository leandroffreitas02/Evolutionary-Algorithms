```python
import numpy as np
import math
from DE import DE
import matplotlib.pyplot as plt
```

```python
A = DE()

names = ['Quadratic Function', 'Sphere Function', 'Rosembrock function', 'Rotated Rastrigin Function', 
         'Rotated Ackley Function', 'Schwefel Function', 'BentCigar Function', 'Ackley Function']

plt.style.use('bmh')
```

# Test 01: F Adapted vs Randomized

```python
for i in range(1, 9):
    fig, ax = plt.subplots()
    plt.title(names[i-1])

    for j in range(10):
        A.unmodified_DE(fun=i, dim=10, max_evaluations=10000)
        ax.semilogy(A.history, color='tab:green')

        A.DE_F(fun=i, dim=10, max_evaluations=10000)
        ax.semilogy(A.history, color='tab:blue')

        A.DE_F(fun=i, dim=10, max_evaluations=10000, randomize=True, prob=1)
        ax.semilogy(A.history, color='tab:red')

    ax.legend(['Normal', 'Adapted', 'Random'], loc='lower left')
    plt.show()
```

# Test 02

```python
for i in range(15):
    fig, ax = plt.subplots()
    plt.title(names[2])

    A.unmodified_DE(fun=3, dim=3, max_evaluations=1000, F=0.4)
    ax.semilogy(A.history, color='tab:green', linewidth=0.8)

    A.unmodified_DE(fun=3, dim=3, max_evaluations=1000, F=0.5)
    ax.semilogy(A.history, color='tab:blue', linewidth=0.8)

    A.unmodified_DE(fun=3, dim=3, max_evaluations=1000, F=0.6)
    ax.semilogy(A.history, color='tab:cyan', linewidth=0.8)

    A.DE_F(fun=3, dim=3, max_evaluations=1000, randomize=True, prob=1)
    ax.semilogy(A.history, color='tab:pink', linewidth=0.8)

    A.DE_F(fun=3, dim=3, max_evaluations=1000, randomize=True, prob=1, random_ub=0.6, random_lb=0.4)
    ax.semilogy(A.history, color='tab:red', linewidth=0.8)

ax.legend(['F=0.4', 'F=0.5', 'F=0.6', 'Random', '0.4-0.6'])
plt.show()
```

# Test 03: Different values for F

```python
for i in range(1, 9):
    fig, ax = plt.subplots()
    plt.title(names[i-1])

    for j in range(15):
        A.unmodified_DE(fun=i, dim=10, max_evaluations=5000, F=0.3)
        ax.semilogy(A.history, color='tab:green', linewidth=0.9)

        A.unmodified_DE(fun=i, dim=10, max_evaluations=5000, F=0.4)
        ax.semilogy(A.history, color='tab:blue', linewidth=0.9)

        A.unmodified_DE(fun=i, dim=10, max_evaluations=5000, F=0.5)
        ax.semilogy(A.history, color='tab:cyan', linewidth=0.9)

        A.DE_F(fun=i, dim=10, max_evaluations=5000, randomize=True, prob=1)
        ax.semilogy(A.history, color='tab:pink', linewidth=0.9)

        A.DE_F(fun=i, dim=10, max_evaluations=5000, randomize=True, prob=1, random_ub=0.6, random_lb=0.4)
        ax.semilogy(A.history, color='tab:red', linewidth=0.9)

    ax.legend(['F=0.3', 'F=0.4', 'F=0.5', 'Random', '0.4-0.6'], loc='lower left')
    plt.show()
```

# Test 04: Randomizing some (Sasa's idea) vs Randomizing all (my idea)

```python
for i in range(1, 9):
    fig, ax = plt.subplots()
    plt.title(names[i-1])

    for j in range(15):
        A.DE_F(fun=i, dim=10, max_evaluations=5000)
        ax.semilogy(A.history, color='tab:green')

        A.DE_F(fun=i, dim=10, max_evaluations=5000, randomize=True, prob=0.1)
        ax.semilogy(A.history, color='tab:red')

        A.DE_F(fun=i, dim=10, max_evaluations=5000, randomize=True, prob=1)
        ax.semilogy(A.history, color='tab:blue')

    ax.legend(['Adapting without randomizing', 'Randomizing 1%', 'Randomizing all'], loc='lower left')
    plt.show()
```

# Random distance 0.1 vs Random distance 0.01

```python
for i in range(1, 9):
    fig, ax = plt.subplots()
    plt.title(names[i-1])

    for j in range(15):
        A.DE_F(fun=i, dim=10, max_evaluations=5000, dist=0.1)
        ax.semilogy(A.history, color='tab:green')

        A.DE_F(fun=i, dim=10, max_evaluations=5000, dist=0.01)
        ax.semilogy(A.history, color='tab:red')

    ax.legend(['0.1', '0.01'], loc='lower left')
    plt.show()
```
